```python
import shutil
import os
from pathlib import Path
```

```python
# Define paths

# Kaggle cache location
kaggle_cache = Path.home() / '.cache' / 'kagglehub' / 'datasets'

# Source paths (datasets in kaggle cache)
dataset_cache = kaggle_cache / 'solesensei' / 'solesensei_bdd100k' / 'versions' / '2'

# Destination: /data/Kaggle dataset/
project_root = Path.cwd().parent.parent
local_dataset_dir = project_root / 'data' / 'custom train data'

print("Path Configuration:")

print(f"Kaggle cache: {kaggle_cache}")
print(f"\nDataset source: {dataset_cache}")
print(f"Exists: {dataset_cache.exists()}")
print(f"\nDestination folder: {local_dataset_dir}")
```

```python
# Create destination folder
local_dataset_dir.mkdir(parents=True, exist_ok=True)
print(f"Created/verified destination folder: {local_dataset_dir}")
```

```python
# Copy Dataset: BDD100K
dataset_dest = local_dataset_dir / 'bdd100k'

if not dataset_cache.exists():
    print("\nSource not found")
elif dataset_dest.exists():
    print(f"\nDestination already exists: {dataset_dest}")
else:
    shutil.copytree(dataset_cache, dataset_dest)
    print(f"\nSuccessfully copied to: {dataset_dest}")
```
