```python
import sys
from pathlib import Path

parent_dir = Path().resolve().parent.parent
sys.path.insert(0, str(parent_dir))

from src.configs import DecordConfig
from src.ingestion.loader import RoadVideoLoader
import torch
import matplotlib.pyplot as plt
import logging
```

```python
config = DecordConfig(video_path=r"D:\Learn\Kaggle Competition\Road-Buddy-Challenge\data\raw\train\videos\0ad45230_447_clip_006_0031_0037_N.mp4")
videoLoaderInstance = RoadVideoLoader(config)
```

```python
frames = [videoLoaderInstance.get_frame(i) for i in range(0, videoLoaderInstance.total_frames // 2, 20)]

num_frames = len(frames)
fig, axes = plt.subplots(1, num_frames, figsize=(15, 10))

if num_frames == 1:
    axes = [axes]

for i, ax in enumerate(axes):
    img_np = frames[i].permute(1, 2, 0).cpu().numpy()
    ax.imshow(img_np)
    ax.axis('off')
    ax.set_title(f"Frame {i*20}")

plt.tight_layout()
plt.show()
```
